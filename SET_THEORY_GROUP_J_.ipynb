
### <center><font color='#4169E1'>1. UNION OPERATOR</font></center>
Combines table rows from two or more different queries into one result

Syntax for union operator
```sql
SELECT column(s) FROM table 1
UNION 
SELECT column(s) FROM table 2 

``` 


#### <font color='#7B68EE'> **i. Union** </font>
---
It does not include duplicates from the two tables.

Example: A union of customer and supplier tables with columns city and region


````python
# A preview of the first 10 columns of customer table where region is 'Western Europe'
%%sql
SELECT City, Region FROM Customer 
WHERE Region = 'Western Europe'
ORDER BY City
LIMIT 10
````

````python
# A preview of the first 10 columns of Supplier table where region is 'Western Europe'
%%sql
SELECT City, Region FROM Supplier
WHERE Region = 'Western Europe'
ORDER BY City
LIMIT 10; 
````

````python
%%sql
SELECT City, Region FROM Customer
WHERE Region='Western Europe'
UNION
SELECT City, Region FROM Supplier
WHERE Region='Western Europe'
ORDER BY City
LIMIT 20; 
````

#### <font color='#7B68EE'> **ii. Union ALL** </font>
---
It  include duplicates from the two tables.

Example: A union all of employee and supplier tables with columns city, region and country

````python
#  A preview of the first 10 records of Employee table
%%sql 
SELECT city, region, country FROM Employee
ORDER BY City
LIMIT 10
````

````python
#  A preview of the first 10 records of Supplier table
%%sql 
SELECT city, region, country FROM Supplier
ORDER BY City
LIMIT 10
````

````python
%%sql 
SELECT city, region, country FROM Employee
UNION ALL 
SELECT city, region, country FROM Supplier
ORDER BY City
LIMIT 35; 
````

````python
%%sql 
# If we replace union all with union, we can see that the duplicate records of London City has been excluded
SELECT city, region, country FROM Employee
UNION  
SELECT city, region, country FROM Supplier
ORDER BY City
LIMIT 20;
````

### <center><font color='#4169E1'>2. INTERSECT OPERATOR</font></center>
It is used to abtain unique records/rows that are common between two tables.

Syntax for intersect operator
```sql
SELECT column(s) FROM table 1
INTERSECT
SELECT column(s) FROM table 2 

```

Example 1: We can get countries where there is a correlation between the customers and employees. 

Example 2: Getting all companies that are both suppliers and shippers 

````python
# A preview of country column from employee table
%%sql 
SELECT Country FROM Employee
````

````python
# A preview of country column from customer table
%%sql 
SELECT Country FROM Customer
````

````python
# Example 1
%%sql 

SELECT Country FROM Customer
INTERSECT 
SELECT Country FROM Employee
````

````python
#  Example 2
%%sql 
SELECT CompanyName AS 'Company' FROM Supplier
INTERSECT 
SELECT CompanyName FROM Shipper
LIMIT 10;
# There is no company that can supply items and do shipping services
````

### <center><font color='#4169E1'>3. EXCEPT OPERATOR</font></center>
EXCEPT returns those tuples that are returned by the first SELECT operation, and not returned by the second SELECT operation

Syntax for except operator
```sql
SELECT column(s) FROM table 1
EXCEPT
SELECT column(s) FROM table 2 
```
Example 1: Getting addresses of places where there is a customer but no employee. It helps the company to know a potential niche for starting a branch

Example 2: Getting all companies that are only suppliers but not shippers

````python
# Example 1
%%sql 

SELECT Postalcode, region AS 'Regional area', city 
FROM Customer
EXCEPT
SELECT Postalcode, region, city 
FROM Employee
ORDER BY City
LIMIT 10; 
````

````python
# Example 2
%%sql
SELECT CompanyName FROM Supplier
EXCEPT
SELECT CompanyName FROM Shipper
LIMIT 10
````
